```kotlin
fun collatz(start: Int) {
    var n = start

    while (true) {
        printNumber(n)
        if (n <= 1) return // This may never exit...
        if (n % 2 == 0) {
            n /= 2
        } else {
            n = 3 * n + 1
        }
    }
}

// This is completely unnecessary, just showcasing some callstack
fun printNumber(int: Int) {
    val str = int.toString()
    println(str)
}
```
